**Análisis estadístico descriptivo del conjunto de datos Titanic**

El conjunto de datos Titanic contiene información sobre los pasajeros del Titanic, incluidos sus nombres, edades, sexos, clases, tarifas y si sobrevivieron o no al hundimiento.

**Conclusión 1: Los pasajeros de primera clase tenían más probabilidades de sobrevivir que los de segunda y tercera clase.**

El gráfico de barras de supervivencia por clase muestra que el 62,5% de los pasajeros de primera clase sobrevivieron, en comparación con el 47,3% de los pasajeros de segunda clase y el 24,2% de los pasajeros de tercera clase.

```
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
titanic = sns.load_dataset("titanic")

# Crear un gráfico de barras de supervivencia por clase
ax = sns.barplot(x="pclass", y="survived", data=titanic)

# Añadir anotaciones en cada barra
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points')

# Mostrar el gráfico
plt.show()
```

**Conclusión 2: Las mujeres tenían más probabilidades de sobrevivir que los hombres.**

El gráfico de barras de supervivencia por sexo muestra que el 74,2% de las mujeres sobrevivieron, en comparación con el 20,6% de los hombres.

```
# asitido por Chatgpt mostrar valores de las barras
# Crear un gráfico de barras de supervivencia por sexo
sns.barplot(x="sex", y="survived", data=titanic)

# Mostrar el gráfico
plt.show()
```

**Conclusión 3: Los pasajeros más jóvenes tenían más probabilidades de sobrevivir.**

El histograma de edades muestra que la mayoría de los pasajeros tenían entre 20 y 40 años. La mediana de edad de los pasajeros que sobrevivieron fue de 28,5 años, en comparación con 25,7 años para los pasajeros que no sobrevivieron.

```
# Crear un histograma de edades
sns.histplot(x="age", data=titanic)

# Mostrar el gráfico
plt.show()
```

**Conclusiones**

Los hallazgos del análisis estadístico descriptivo del conjunto de datos Titanic son los siguientes:

* Los pasajeros de primera clase tenían más probabilidades de sobrevivir que los pasajeros de segunda y tercera clase.
* Las mujeres tenían más probabilidades de sobrevivir que los hombres.
* Los pasajeros más jóvenes tenían más probabilidades de sobrevivir.

Estos hallazgos sugieren que el sexo, la clase y la edad eran factores importantes que determinaron la supervivencia de los pasajeros del Titanic.
